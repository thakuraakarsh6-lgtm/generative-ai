```python
#default argument -- these argument value during defined a function
def area_perimeter(width = 2, height = 4):
  area = width*height
  perimerter = 2*(width+height)
  return area , perimerter
```

```python
def sum_number (*args):
  sum = 0
  for num in args:
    sum+=num
  return sum
```

```python
#keyboard variable lenght argument behave a dictionary
def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
```

```python
test(a = 10, b = 20, c = 30)
```

```python
#keyboard variable lenght argument behave a dictionary
def pass_students(**kwargs):
  pass_student_list = []
  for name, marks in kwargs.items():
    if marks > 33:
      pass_student_list.append(name)
  return pass_student_list
```

```python
pass_students(A= 40, B= 30, C= 60)
```
