```python
# bar polt -> it is work on numerical as well as catagorical data
import matplotlib.pyplot as plt
import numpy as np
friut = np.array([",Apple","Banana", "orange","watermelon"])
quintity = np.array([100,200,150,120])
color = np.array(["Red","yellow","orange","green"])
plt.bar (friut,quintity,color = color)
plt.show()
```
