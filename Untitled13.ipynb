```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Create sample data
data = {'Marks' : [85, 90, 78, 92, 88],
        'Hours' : [5, 6, 4, 7, 5]}
df = pd.DataFrame(data)
print(df.describe())
print("Correlation between marks & Hours:",
       df['Marks'].corr(df['Hours']))
plt.plot(df['Hours'], df['Marks'],
marker = 'o')
plt.xlabel('Hours')
plt.ylabel('Marks')
plt.show()
```
